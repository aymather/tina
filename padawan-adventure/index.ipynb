```python
import matplotlib.pyplot as plt
import pandas as pda
```

```python
df=pda.read_csv("diamond.csv")
print(df)
```

```python
df.plot(x='x', y='y')
plt.show()
```

```python
mean_value_x = df['x'] .mean()
print ("mean_of_x:", mean_value_x)
```

```python
median_value_x = df ['x'] .median()
print ("median_of_x:", median_value_x)
```

```python
mode_value_x = df ['x'] .mode()
print("mode_of_x:\n", mode_value_x)
```

```python
std_deviation_x = df['x'].std()
print("standard_deviation_of_x:", std_deviation_x)
```

```python
mean_value_y = df['y'] .mean()
print("mean_of_y:", mean_value_y)
```

```python
median_value_y = df['y'] .median()
print("median_of_y:", median_value_y)
```

```python
mode_value_y = df['y'] .mode()
print('mode_of_y:\n', mode_value_y)
```

```python
std_deviation_y = df['y'] .std()
print("standard_deviation_of_y:", std_deviation_y)
```

```python
min_x = df["x"] .min()
print("min_x", min_x)
```

```python
max_x = df["x"] .max()
print("max_of_x", max_x)
```

```python
min_y = df["y"] .min()
print("min_y", min_y)
```

```python
max_y = df["y"] .max()
print("max_y", max_y)
```
